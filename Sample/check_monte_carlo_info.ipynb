# Sherpa

```python
import pyhepmc
from tqdm import tqdm

def calculate_polarization_cross_sections(hepmc_file_path):
    """
    從 Sherpa HepMC3 檔案中提取極化權重，並計算 LL, LT, TT 的物理截面。
    """
    # 初始化權重累加器
    sum_weights = {
        'total_nominal': 0.0,
        'LL': 0.0,
        'LT': 0.0,       # 包含 LT 與 TL
        'TT': 0.0,
        'interference': 0.0 # 干涉項
    }
    
    total_xsec_pb = 0.0
    
    print(f"Reading and processing {hepmc_file_path} ...")
    
    # 使用 with 確保資源正確釋放，pyhepmc.open 會逐事件讀取，避免 OOM
    with pyhepmc.open(hepmc_file_path) as f:
        for i, event in tqdm(enumerate(f)):
            if not event:
                break
                
            # 從第一筆事件中提取 Sherpa 記錄的整體截面 (Cross Section)
            # HepMC3 中預設單位置為 pb
            if i == 0 and event.cross_section:
                total_xsec_pb = event.cross_section.xsec(0)
                
            # 將 weight_names 與當前事件的 weights 組合為 Dictionary 方便查找
            weight_dict = dict(zip(event.weight_names, event.weights))
            
            # 提取標稱總權重 (Nominal)
            w_nom = weight_dict.get('Weight', 0.0)
            
            # 提取 COM 座標系下的極化權重 (針對 W+ W+)
            # 若您後續處理 W- W-，請將 'W+.0' 改為 'W-.0'
            w_LL = weight_dict.get('PolWeight_COM.W+.0_W+.0', 0.0)
            
            # LT 態應包含 W1(L)W2(T) 與 W1(T)W2(L) 兩種組合
            w_LT = weight_dict.get('PolWeight_COM.W+.0_W+.T', 0.0) + \
                   weight_dict.get('PolWeight_COM.W+.T_W+.0', 0.0)
                   
            w_TT = weight_dict.get('PolWeight_COM.W+.T_W+.T', 0.0)
            w_int = weight_dict.get('PolWeight_COM.int', 0.0)
            
            # 累加至總和
            sum_weights['total_nominal'] += w_nom
            sum_weights['LL'] += w_LL
            sum_weights['LT'] += w_LT
            sum_weights['TT'] += w_TT
            sum_weights['interference'] += w_int

    # 驗證權重守恆: Nominal 應該近似等於 LL + LT + TT + 干涉項
    sum_components = sum_weights['LL'] + sum_weights['LT'] + sum_weights['TT'] + sum_weights['interference']
    ratio_check = sum_components / sum_weights['total_nominal'] if sum_weights['total_nominal'] > 0 else 0
    
    print("\n" + "="*50)
    print("Weight Conservation Check:")
    print(f"  Sum of Components / Nominal = {ratio_check:.4f} (Should be close to 1.0)")
    
    # 計算各極化態的絕對截面 (單位: pb)
    print("\nCross-Section Results:")
    print(f"  Total Cross-Section (Sherpa): {total_xsec_pb:.6e} pb")
    
    xsec_results = {}
    for pol in ['LL', 'LT', 'TT']:
        # 利用權重比例計算該極化態的截面
        fraction = sum_weights[pol] / sum_weights['total_nominal']
        xsec = total_xsec_pb * fraction
        xsec_results[pol] = xsec
        
        # 轉換為 fb (femtobarn) 顯示，這在 VBS 中較為直觀
        print(f"  {pol} Fraction: {fraction*100:.2f}% | Cross-Section: {xsec*1000:.4f} fb")
        
    print("="*50)
    
    return xsec_results

# 執行範例：
results = calculate_polarization_cross_sections("Sherpa/EW_WpWpjj/hepmc_data/vbs_wpwp_seed1.hepmc")
```
